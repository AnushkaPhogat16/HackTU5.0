```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv(r'data/RIC_sim1.csv')
```

```python
plt.scatter(data['longitude'],data['latitude'],c='red',marker='o',label = 'location')
plt.grid(True)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Latitude and Longitude')
plt.legend()
plt.show()
```

```python
# optimal number of clusers
wcss = [] # within clusteers sum of square
n_range = 15
for i in range(1,n_range): # trying at max 20 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,n_range),wcss,'bx-')
plt.title('the elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, init='k-means++',random_state=42)
kmeans.fit(data)

#img = plt.imread(r'img/map.png')
#plt.imshow(img)
plt.scatter(data['latitude'],data['longitude'], c=kmeans.labels_)
plt.grid(True)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Clustering Location as per Data')
plt.legend()
plt.show()
```
